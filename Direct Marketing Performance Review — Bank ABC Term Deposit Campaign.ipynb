Tidak ditemukan missing value didalam dataset ini.

Beberapa kolom kategorikal masih memiliki label **"unknown"** yang merepresentasikan (missing value) yang telah diimputasi sebagai kategori tersendiri. Untuk memperjelas kualitas data dan potensi bias analisis, perlu dihitung proporsi nilai **"unknown"** pada masing-masing kolom sebelum melanjutkan tahap analisis eksploratif.


```python
# Mengambil hanya kolom kategorikal (tipe object)
cat_cols = df.select_dtypes(include='object').columns.tolist()
cat_cols
```

```python
# Menghitung jumlah dan proporsi nilai "unknown" pada setiap kolom kategorikal
unknown_stats = pd.DataFrame(index=cat_cols, columns=['count_unknown', 'prop_unknown'])

for col in cat_cols:
    count_unknown = (df[col] == 'unknown').sum()
    prop_unknown = count_unknown / len(df)  # proporsi (0–1)
    
    unknown_stats.loc[col, 'count_unknown'] = count_unknown
    unknown_stats.loc[col, 'prop_unknown'] = prop_unknown

# Mengubah proporsi jadi persen dan merapikan tampilan
unknown_stats['prop_unknown'] = (unknown_stats['prop_unknown'].astype(float) * 100).round(2)

# Urut dari proporsi unknown terbesar
unknown_stats.sort_values('prop_unknown', ascending=False)

unknown_stats.style.format({'prop_unknown': '{:.2f}%'})

```

Berdasarkan perhitungan proporsi nilai **"unknown"** pada kolom kategorikal, diperoleh hasil sebagai berikut:

- `default`  : 8.597 baris (≈ **20,87%**)
- `education`: 1.731 baris (≈ **4,20%**)
- `housing`  : 990 baris (≈ **2,40%**)
- `loan`     : 990 baris (≈ **2,40%**)
- `job`      : 330 baris (≈ **0,80%**)
- `marital`  : 80 baris (≈ **0,19%**)

Sementara itu, kolom `contact`, `month`, `day_of_week`, `poutcome`, dan `y` tidak memiliki nilai **"unknown"**.

Proporsi **"unknown"** yang cukup besar pada variabel `default` menunjukkan bahwa informasi status gagal bayar historis nasabah banyak yang tidak terdokumentasi dengan baik. Meskipun proporsi nilai **"unknown"** pada variabel tersebut cukup besar, pada analisis eksploratif ini kolom `default` **tidak disertakan** dalam rangkaian analisis utama yang berfokus pada **investasi dan efektivitas kampanye pemasaran deposito berjangka**. Dengan demikian, nilai **"unknown"** pada `default` tidak diberikan perlakuan khusus lebih lanjut. Keputusan ini diambil agar fokus analisis tetap pada variabel yang lebih relevan untuk pengambilan keputusan kampanye, sekaligus menghindari potensi bias akibat kualitas data `default` yang kurang memadai.

Untuk variabel kategorikal selain `default`, proporsi nilai **"unknown"** berada di bawah **5%**. Dalam praktik analisis data, tingkat kehilangan data di bawah ambang ini umumnya tidak terlalu mengganggu gambaran agregat, sehingga nilai **"unknown"** masih dapat dipertahankan sebagai kategori tersendiri tanpa perlu perlakuan imputasi yang agresif.

Meskipun demikian, untuk meminimalkan potensi bias, tetap penting untuk memahami mekanisme hilangnya data tersebut: apakah kemunculan **"unknown"** cenderung bersifat **MCAR (Missing Completely at Random)**, **MAR (Missing at Random)**, atau justru **MNAR (Missing Not at Random)**. Oleh karena itu, pada tahap analisis berikutnya perlu ditelusuri apakah kategori **"unknown"** pada variabel-variabel tersebut memiliki keterkaitan dengan kolom lain.

Jika distribusi **"unknown"** relatif merata di berbagai kelompok, asumsi MCAR/MAR masih cukup masuk akal dan risiko bias menjadi kecil. Sebaliknya, bila **"unknown"** terkonsentrasi pada segmen tertentu, maka hasil analisis yang melibatkan variabel terkait perlu diinterpretasikan dengan lebih hati-hati.



```python
# Kolom dengan unknown > 0% dan < 5%
low_unknown_cols = unknown_stats[
    (unknown_stats['prop_unknown'] > 0) & (unknown_stats['prop_unknown'] < 5)
].index.tolist()

low_unknown_cols
```

Cek keterkaitan kolom di atas yang memiliki value “unknown” dengan target kolom `y` (uji chi-square). 

```python
# Menghitung uji chi square antara unknown pada kolom tertentu dengan target 'y'
from scipy.stats import chi2_contingency

for col in low_unknown_cols:
    # flag apakah baris tsb unknown atau tidak
    flag_unknown = df[col].eq('unknown')
    
    # tabulasi silang unknown vs y
    ct = pd.crosstab(flag_unknown, df['y'])
    
    chi2, p, dof, expected = chi2_contingency(ct)
    
    print(f"\n=== Kolom: {col} ===")
    print("Crosstab unknown vs y:")
    print(ct)
    print(f"p-value chi-square: {p:.4f}")
```

p-value > 0.05 → distribusi antara baris unknown dan non-unknown tidak berbeda signifikan → missing lebih dekat ke MCAR/MAR

p-value ≤ 0.05 → ada perbedaan signifikan → menunjukkan bahwa siapa yang “unknown” cenderung berasal dari segmen tertentu → missing tidak sepenuhnya acak/ ada pattern tertentu. 

**Maka dengan menggunakan Uji chi-square: apakah label **"unknown"** berkaitan dengan respon investasi (`y`)?**

Kita telah mengecek apakah nilai yang sudah di-*impute* sebagai **"unknown"** bersifat acak atau justru punya pola tertentu terhadap respon investasi, sehingga telah dilakukan uji **chi-square** antara:

- flag_unknown (True/False)  ×  `y` (no/yes)

untuk beberapa variabel kategorikal: `job`, `marital`, `education`, `housing`, dan `loan`.



**Ringkasan hasil**

- **`job`**
  - Tingkat konversi pada baris **known** dan **unknown** hampir sama (≈11,2–11,3%).  
  - **p-value ≈ 1,000** → **tidak ada bukti** bahwa keberadaan label `job = unknown` berkaitan dengan perbedaan peluang `y = yes`. Maka unknown di kolom Job adalah Missing at Completely at Random (MCAR).

- **`marital`**
  - Konversi `marital = unknown` sedikit lebih tinggi, tetapi jumlah kasusnya sangat kecil (hanya 80 baris).  
  - **p-value ≈ 0,38** → **secara statistik tidak signifikan**; perbedaan ini dapat dianggap kebetulan sampel. Maka unknown di Marital hampir mendekati Missing Completely at Random (MCAR). 

- **`education`**
  - Baris dengan **`education = unknown`** menunjukkan **conversion rate yang lebih tinggi** dibandingkan baris dengan data pendidikan lengkap (sekitar **14,5% vs 11%**).
  - Hasil **uji Chi-Square** menghasilkan **p-value ≈ 0,0000**, yang menunjukkan **bukti statistik yang sangat kuat** adanya hubungan antara status `education = unknown` dan respons terhadap penawaran investasi.
  - Temuan ini mengindikasikan bahwa *missing value* pada kolom **`education`** **tidak bersifat acak terhadap variabel target (`y`)**, melainkan memiliki pola tertentu.
  - Dengan demikian, mekanisme *missing value* pada **`education`** **tidak dapat dikategorikan sebagai MCAR**, dan lebih tepat diklasifikasikan sebagai **Missing Not At Random (MNAR)**.


- **`housing`**
  - Perbedaan **conversion rate** antara baris dengan data lengkap dan **`housing = unknown`** relatif sangat kecil (sekitar **11,6% vs 10,8%**).
  - Hasil **uji Chi-Square** menunjukkan **p-value ≈ 0,68**, yang mengindikasikan **tidak adanya hubungan signifikan** antara status missing pada `housing` dan respons terhadap investasi (`y`).
  - Dengan demikian, *missing value* pada kolom **`housing`** dapat dikategorikan sebagai **Missing Completely At Random (MCAR)**.

- **`loan`**
  - Pola yang diamati serupa dengan kolom **`housing`**, di mana **conversion rate** antara data lengkap dan **`loan = unknown`** hampir tidak berbeda.
  - Nilai **p-value ≈ 0,68** menunjukkan **tidak terdapat hubungan statistik yang signifikan** antara missing value pada `loan` dan variabel target (`y`).
  - Oleh karena itu, *missing value* pada kolom **`loan`** juga dapat diklasifikasikan sebagai **Missing Completely At Random (MCAR)**.


**Implikasi untuk perlakuan data**

- Untuk **`job`, `marital`, `housing`, dan `loan`**, label **"unknown"** tidak menunjukkan perbedaan signifikan terhadap peluang `y = yes`.  
  ➜ Dalam konteks respon investasi, *missing value* pada variabel-variabel ini bisa diperlakukan sebagai **kategori biasa** (dibiarkan sebagai `unknown`) tanpa penyesuaian khusus, karena tidak menimbulkan bias yang jelas terhadap `y`.

- Untuk **`education`**, kondisi berbeda:
  - `education = unknown` justru sedikit **lebih responsif** terhadap kampanye.
  - Karena proporsinya masih di bawah 5% tetapi secara statistik berbeda, segmen ini **perlu dicatat secara eksplisit dalam interpretasi** (bukan dihapus atau digabung sembarang).  
  ➜ Pada tahap analisis, setiap kali membahas pola berdasarkan pendidikan, kategori `education = unknown` akan ditampilkan dan dijelaskan sebagai kelompok dengan pencatatan data yang bias atau kurang lengkap. 

Secara keseluruhan, uji chi-square ini mendukung keputusan bahwa **mayoritas nilai "unknown" dapat dipertahankan sebagai kategori tersendiri tanpa proses imputasi lanjutan**.

```python
# Menghitung jumlah outlier pada setiap kolom numerik menggunakan IQR

def count_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = series[(series < lower_bound) | (series > upper_bound)]
    return outliers.count()
outlier_counts = {}
for col in numeric_cols:
    outlier_counts[col] = count_outliers(df[col])
outlier_counts

```

### **5.2 Analisa Convertion Rate Berdasarkan Segmentasi Profiling**

#### **5.2.1 Convertion Rate Berdasarkan Variabel `age`**

```python
# Mencari conversion rate rate pada kolom age berdasarkan kolom y
age_conversion = df.groupby('age')['y'].value_counts(normalize=True).unstack()
age_conversion['yes'] = age_conversion['yes'] * 100
age_conversion['no'] = age_conversion['no'] * 100
#round by 2 decimal
age_conversion = age_conversion.round(2)
age_conversion  
```

Untuk mempermudah analisa, akan dilakukan grouping untuk kolom `age` menjadi 5 kategori umur yaitu `Early Working Age 18 - 28`,`Prime Working Age 29 - 40`,`Late Working Age 41 - 52`, `Pre-Retirement 53 - 64`, dan `Retirement 65+`. 

```python
# Hitung total convert (yes), n (no), dan conversion rate per age group

df['age_group'] = df['age'].map(
    lambda age: (
        'Early Working Age 18 - 28' if age < 29 else
        'Prime Working Age 29 - 40' if age < 41 else
        'Late Working Age 41 - 52' if age < 53 else
        'Pre-Retirement 53 - 64' if age < 65 else
        'Retirement 65+'
    )
)
# use y == 'yes' as the conversion indicator because there is no 'conv' column
conversions_by_age_group = df.groupby('age_group')['y'].apply(lambda s: (s == 'yes').sum())
total_by_age_group = df.groupby('age_group')['y'].count()

conversion_rate_by_age_group = (conversions_by_age_group / total_by_age_group * 100).round(2)

conversion_df = pd.DataFrame({
    'Total (all)': total_by_age_group,
    'Total Converts (yes)': conversions_by_age_group,
    'Total no': total_by_age_group - conversions_by_age_group,
    'Conversion Rate (%)': conversion_rate_by_age_group
}).sort_index()

display(conversion_df)
```

```python
# Visualize conversion rate by age group
plt.figure(figsize=(10, 6)) 
sns.barplot(x=conversion_df.index, y='Conversion Rate (%)', data=conversion_df, palette='magma')
plt.title("Conversion Rate by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Conversion Rate (%)")
plt.xticks(rotation=45)
plt.show()
```

Tabel di atas menunjukkan hubungan antara **kelompok usia nasabah (`age_group`)** dan keputusan untuk mengambil **investasi berjangka**. Secara umum, **conversion rate keseluruhan** berada di kisaran **±11%**, sehingga kelompok usia dengan nilai di atas angka ini dapat dianggap **lebih responsif dibandingkan rata-rata populasi**.

#### **5.2.2 Convertion Rate Berdasarkan Variabel `job`**

```python
# Ringkasan konversi kategori 'Job'
def summarize_conversion(col):
    total_customers = len(df)
    total_yes = (df['y'] == 'yes').sum()
    
    summary = (
        df.groupby(col)['y']
          .agg(
              total_n='size',
              n_yes=lambda x: (x == 'yes').sum(),
              n_no =lambda x: (x == 'no').sum(),
          )
    )
    
    summary['rate_yes_%']  = (summary['n_yes'] / summary['total_n'] * 100).round(2)
    summary['rate_no_%']   = (summary['n_no']  / summary['total_n'] * 100).round(2)
    summary['share_of_all_%'] = (summary['total_n'] / total_customers * 100).round(2)
    summary['share_of_yes_%'] = (summary['n_yes'] / total_yes * 100).round(2)
    
    # urut dari kategori paling “responsif”
    summary = summary.sort_values('rate_yes_%', ascending=False)
    return summary

summarize_conversion('job')
job_conv = summarize_conversion("job")
job_conv
```

```python
# Ringkasan konversi per kategori dari satu kolom
def summarize_conversion_local(data, col):
    total_customers = len(data)
    total_yes = (data['y'] == 'yes').sum()
    
    summary = (
        data.groupby(col)['y']
            .agg(
                total_n='size',
                n_yes=lambda x: (x == 'yes').sum(),
                n_no =lambda x: (x == 'no').sum(),
            )
    )
    
    summary['rate_yes_%']      = (summary['n_yes'] / summary['total_n'] * 100).round(2)
    summary['rate_no_%']       = (summary['n_no']  / summary['total_n'] * 100).round(2)
    summary['share_seg_%']     = (summary['total_n'] / total_customers * 100).round(2)
    summary['share_yes_seg_%'] = (summary['n_yes'] / total_yes * 100).round(2)
    
    return summary.sort_values('rate_yes_%', ascending=False)
```

#### **5.6.1 Convertion Rate Berdasarkan Variabel `contact`**

```python
# Ringkasan konversi kategori 'Contact'
def summarize_conversion(col):
    total_customers = len(df)
    total_yes = (df['y'] == 'yes').sum()
    
    summary = (
        df.groupby(col)['y']
          .agg(
              total_n='size',
              n_yes=lambda x: (x == 'yes').sum(),
              n_no =lambda x: (x == 'no').sum(),
          )
    )
    
    summary['rate_yes_%']  = (summary['n_yes'] / summary['total_n'] * 100).round(2)
    summary['rate_no_%']   = (summary['n_no']  / summary['total_n'] * 100).round(2)
    summary['share_of_all_%'] = (summary['total_n'] / total_customers * 100).round(2)
    summary['share_of_yes_%'] = (summary['n_yes'] / total_yes * 100).round(2)
    
    # urut dari kategori paling “responsif”
    summary = summary.sort_values('rate_yes_%', ascending=False)
    return summary

summarize_conversion('contact')
```

```python
# Visualisasi urutan distribusi jumlah yes/no per jenis kontak (contact) dengan stack bar
contact_conv = summarize_conversion("contact")
contact_sorted = contact_conv.sort_values('rate_yes_%', ascending=True) 
df_stack = contact_sorted
x = np.arange(len(df_stack.index))
plt.figure(figsize=(6, 3))
plt.bar(x, df_stack['n_no'],  label='No (tidak investasi)')
plt.bar(x, df_stack['n_yes'], bottom=df_stack['n_no'], label='Yes (investasi)')
plt.xticks(x, df_stack.index, rotation=0)
plt.ylabel('Jumlah nasabah')
plt.title('Distribusi jumlah Yes/No investasi per jenis kontak')
plt.legend()
plt.tight_layout()
plt.show()
```

**Investasi berdasarkan `contact`**

- **`cellular`** menyumbang **mayoritas nasabah yang berinvestasi (`yes`)** dan menunjukkan efektivitas channel yang paling tinggi.
- **`telephone`** memiliki jumlah konversi yang jauh lebih rendah dan didominasi oleh nasabah yang tidak berinvestasi (`no`).

Kontak melalui **`cellular`** merupakan **channel utama yang paling efektif** untuk mendorong investasi berjangka, sedangkan **`telephone`** berperan sebagai channel pendukung dan kurang optimal jika diprioritaskan.

Conversion rate per month, day_of_week, dan contact

```python
# Fungsi ringkasan konversi per month, day_of_week, contact 
def conv_table(col):
    out = (df.groupby(col)["y_bin"]
             .agg(n="size", yes="sum", conversion="mean")
             .sort_values("conversion", ascending=False))
    return out

month_conv   = conv_table("month")
dow_conv     = conv_table("day_of_week")
contact_conv = conv_table("contact")

month_conv, dow_conv, contact_conv

```

```python
# Siapkan data: buang bulan tanpa kontak
month_plot = month_conv[month_conv["n"] > 0].reset_index()

# Definisikan urutan bulan
month_order = ["jan", "feb", "mar", "apr", "may", "jun",
               "jul", "aug", "sep", "oct", "nov", "dec"]

# Jadikan 'month' sebagai kategori berurutan, lalu sort
month_plot["month"] = pd.Categorical(
    month_plot["month"],
    categories=month_order,
    ordered=True
)
month_plot = month_plot.sort_values("month")

plt.figure(figsize=(10, 5))

# Sumbu kiri: bar jumlah kontak
ax1 = plt.gca()
ax1.bar(month_plot["month"], month_plot["n"],
       color="lightgray", label="Jumlah kontak (N)")
ax1.set_ylabel("Jumlah nasabah yang dihubungi (N)")
ax1.set_xlabel("Bulan kampanye")

# Sumbu kanan: garis conversion rate
ax2 = ax1.twinx()
ax2.plot(
    month_plot["month"],
    month_plot["conversion"] * 100,
    marker="o",
    linewidth=2,
    label="Conversion rate",
)
ax2.set_ylabel("Conversion rate (%)")

plt.title("Volume kontak vs Conversion rate per bulan")
plt.xticks(rotation=45)

# Gabungkan legend
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc="upper right")

plt.tight_layout()
plt.show()
```

```python
# Visualisasi conversion rate dengan label pada bar
def plot_conv_bar_with_labels(table, title, xlabel):
    t = table.reset_index().copy()
    cat = t.columns[0]

    plt.figure(figsize=(10,4))
    bars = plt.bar(t[cat].astype(str), t["conversion"])
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel("Conversion rate")
    plt.xticks(rotation=45, ha="right")

    for b, v in zip(bars, t["conversion"]):
        plt.text(b.get_x() + b.get_width()/2, b.get_height(), f"{v:.1%}",
                 ha="center", va="bottom", fontsize=9)

    plt.tight_layout()
    plt.show()

plot_conv_bar_with_labels(month_conv, "Conversion Rate by Month", "Month")
plot_conv_bar_with_labels(dow_conv, "Conversion Rate by Day of Week", "Day of Week")
plot_conv_bar_with_labels(contact_conv, "Conversion Rate by Contact Channel", "Contact")
```

- Rata-rata durasi untuk nasabah **`no` hanya ~221 detik** (±3,5 menit), sementara untuk nasabah **`yes` mencapai ~553 detik** (±9 menit). Artinya, secara rata-rata **agent menghabiskan waktu lebih dari 2× lebih lama** pada panggilan yang akhirnya berujung investasi.
- Median durasi `no` berada di sekitar **163 detik**, sedangkan median `yes` di sekitar **449 detik**. Dengan kata lain, **50% panggilan yang berhasil konversi berlangsung ≥ 7–8 menit**, sementara separuh panggilan yang gagal konversi berhenti di kisaran 2–3 menit.
- Kuartil bawah dan atas juga bergeser ke kanan untuk kelompok `yes` (Q1: 253 detik, Q3: 741 detik) dibanding `no` (Q1: 95 detik, Q3: 279 detik), yang mengindikasikan bahwa **sebagian besar interaksi yang produktif memang membutuhkan waktu obrolan yang lebih panjang**.

Polanya terlihat jelas pada boxplot: distribusi durasi untuk label `yes` bergeser ke kanan (lebih lama) dan memiliki “tubuh” box yang lebih tinggi dibanding `no`. Hal ini mendukung kesimpulan bahwa:

> **Durasi panggilan yang lebih panjang berasosiasi kuat dengan peluang konversi yang lebih tinggi.**  
> Panggilan yang terlalu singkat cenderung tidak memberikan waktu cukup bagi agent untuk menjelaskan manfaat produk dan menjawab keberatan nasabah.

Namun demikian, adanya outlier durasi panjang pada label `no` menunjukkan bahwa **durasi panjang bukan satu-satunya syarat**; kualitas percakapan dan kecocokan segmen juga menentukan. Durasi bisa dianggap sebagai **pra-syarat minimal** (harus cukup lama agar edukasi terjadi), tetapi **tidak menjamin konversi** tanpa didukung skrip komunikasi dan segmentasi yang tepat.


Selanjutnya bagi kategori durasi (bins) untuk analisis konversi per rentang

```python
# Kategorikan duration ke dalam beberapa bin
bins   = [0, 100, 300, 600, np.inf]
labels = ['<=100s', '101-300s', '301-600s', '>600s']

df['duration_cat'] = pd.cut(
    df['duration'],
    bins=bins,
    labels=labels,
    include_lowest=True,
    right=True
)

df['duration_cat'].value_counts()
```

```python
# Ringkasan konversi per kategori duration_cat
duration_summary = (
    df.groupby('duration_cat')['y']
      .agg(
          total_n = 'size',
          n_yes   = lambda x: (x == 'yes').sum(),
          n_no    = lambda x: (x == 'no').sum()
      )
)

duration_summary['conversion_pct'] = (
    duration_summary['n_yes'] / duration_summary['total_n'] * 100
).round(2)

duration_summary
```

```python
# Visualisasi jumlah kontak dan conversion rate per kategori duration_cat
duration_plot = duration_summary.reset_index()

plt.figure(figsize=(8,4))
ax1 = plt.gca()

# Bar: jumlah nasabah
ax1.bar(duration_plot['duration_cat'], duration_plot['total_n'],
        alpha=0.4, label='Jumlah nasabah (N)')
ax1.set_xlabel('Kategori durasi panggilan')
ax1.set_ylabel('Jumlah nasabah (N)')

# Axis kedua: conversion rate
ax2 = ax1.twinx()
ax2.plot(duration_plot['duration_cat'], duration_plot['conversion_pct'],
         marker='o', linewidth=2, label='Conversion rate (%)')
ax2.set_ylabel('Conversion rate (%)')

plt.title('Jumlah kontak dan conversion rate per kategori durasi')
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper right')

plt.tight_layout()
plt.show()
```

Data di atas menunjukkan hubungan yang **sangat kuat** antara durasi panggilan dan peluang konversi:

- Panggilan **sangat singkat (≤100 detik)** jumlahnya cukup banyak (±10.000 call), tetapi hampir tidak menghasilkan penjualan: conversion rate hanya **0,81%**. Ini tipikal *cold call* yang cepat ditolak nasabah.
- Panggilan **101–300 detik** adalah yang paling umum (±20.000 call) dengan conversion rate naik menjadi **7,2%**. Artinya, butuh minimal **1,5–5 menit** agar pesan produk mulai tersampaikan.
- Pada rentang **301–600 detik**, meskipun volume call lebih kecil (7.740), jumlah *yes* hampir sama dengan segmen sebelumnya (1.438 vs 1.437) sehingga conversion rate melonjak ke **18,6%**. Ini menunjukkan bahwa **pembicaraan yang lebih dalam dan dialog dua arah** jauh lebih produktif.
- Segmen **>600 detik** memang paling sedikit (3.464 call), tetapi menghasilkan conversion tertinggi, **48,6%** – hampir **setengah** panggilan super-panjang berakhir dengan pembukaan investasi.

Pola ini menjelaskan bahwa **semakin panjang durasi panggilan, semakin besar peluang konversi**.  
Call yang terlalu singkat cenderung tidak memberi kesempatan bagi agent untuk membangun kepercayaan dan menjelaskan manfaat produk, sementara call yang cukup panjang biasanya terjadi karena nasabah sudah menunjukkan minat, sehingga peluang closing meningkat tajam.

**Implikasi bagi bank:**

1. Menghindari pola call *“asal kontak”* yang berakhir sangat cepat.
2. Mendorong **kualitas percakapan** (skrip yang lebih konsultatif, bukan sekadar membaca penawaran) agar semakin banyak panggilan yang masuk ke rentang **≥300 detik**, di mana tingkat konversi mulai besar.

### 7.2 **Rekomendasi 2 – Optimasi Strategi Komunikasi untuk Segmen Under-Performance**

Berdasarkan EDA, beberapa segmen pasar menunjukkan **konversi di bawah rata-rata** meskipun volumenya besar, antara lain:

- **Usia produktif utama** (29–52 tahun).

---

#### 7.2.1 Prinsip konstruksi penyampaian komunikasi untuk segmen under-performance

  1. **Sederhanakan narasi produk, supaya pemahaman investasi lebih mudah diterima**  
   - Minimalkan istilah teknis (“yield tahunan”, “tenor”, dsb.), ganti dengan bahasa **“uang masuk per bulan”** dan contoh konkret.  
   - Fokus pada 3 pesan inti: 
     - **aman** (risiko rendah),  
     - **tidak ribet** (auto-debit, bisa pantau via mobile banking),  
     - **tujuan jelas** (tabungan pendidikan anak, dana cadangan, dana usaha).

  2. **Sorot manfaat yang “dekat dengan kehidupan sehari-hari”**  
   - Untuk segmen pekerja lapangan / pendidikan menengah, manfaat abstrak jangka panjang kurang “nyangkut”.  
   - Tekankan contoh:  
     - “Jika Bapak/Ibu sisihkan X per bulan, dalam Y tahun bisa terkumpul dana setara [1x motor / 6 bulan gaji cadangan / modal usaha kecil].”

  3. **Ubahlah nada komunikasi dari “jualan produk” menjadi “bantu mengamankan penghasilan”**  
   - Alihkan skrip dari kalimat seperti *“Bank kami menawarkan produk investasi berjangka…”*  
   - Menjadi *“Bagaimana kalau sebagian kecil penghasilan Bapak/Ibu disisihkan supaya tidak habis begitu saja setiap bulan?”*  
   - Pendekatan **konsultatif** lebih cocok untuk segmen yang awalnya tidak terlalu tertarik pada investasi.

---

#### 7.2.2 Strategi komunikasi khusus per kelompok segmen

##### 7.2.2.1 Pendidikan menengah (`middle.school + high.school`, mid–high segment)

  **Problem utama:** kontribusi leads sangat besar, tetapi conversion rate di bawah rata-rata; banyak konversi baru terjadi pada panggilan sangat panjang (mahal dari sisi waktu agent).

  **Arah strategi:**

  - **Turunkan kompleksitas penjelasan produk**  
  - Gunakan analogi yang familiar: *“tabungan yang dikunci supaya tidak ikut terpakai”*.  
  - Tambahkan visual bantu (banner/simulasi sederhana) via WhatsApp/MSMS sebelum atau sesudah call.

  - **Bangun kepercayaan terlebih dahulu**  
  - Awal komunikasi lebih banyak pada: 
    - keamanan bank,  
    - testimoni sederhana,  
    - dan contoh nyata (tanpa menakut-nakuti).  
  - Hindari langsung bicara angka besar; mulai dari nominal kecil yang terasa “masuk akal”.

  - **Batasi durasi edukasi di telepon, lanjutkan lewat kanal lain**  
    - Karena edukasi penuh di satu call mahal, gunakan call untuk trigger minat, lalu kirimkan rangkuman / simulasi melalui SMS/WhatsApp/Email untuk dibaca mandiri.

##### 7.2.2.2 Blue-collar dan Services (`job`)

  **Problem utama:** dua segmen ini banyak dihubungi, tetapi conversion rate paling rendah; indikasi ketidaksesuaian antara cara kita menjual dengan realitas finansial dan preferensi mereka.

  **Arah strategi:**

  - **Tonjolkan fleksibilitas nominal & risiko rendah**  
    - Tawarkan **entry ticket kecil** yang realistis terhadap penghasilan harian/bulanan.  
    - Hindari kesan produk “untuk orang berduit”, tekankan bahwa ini bisa dimulai dari “jumlah yang tidak mengganggu pengeluaran bulanan”.

  - **Gunakan contoh kasus yang relevan dengan pola hidup mereka**  
    - Misal: cadangan dana jika tiba-tiba kontrak kerja tidak diperpanjang; tabungan untuk kebutuhan darurat keluarga; cicilan barang tanpa harus berhutang.

  - **Sesuaikan waktu dan channel**  
    - Hubungi di jam yang tidak mengganggu jam kerja inti (bukan saat pergantian shift).  
    - Kombinasikan dengan pesan singkat (SMS/WhatsApp) yang bisa dibaca ketika mereka senggang, bukan hanya mengandalkan telepon langsung.

##### 7.2.2.3 Usia 29–52 tahun (segmen produktif utama)

  **Problem utama:** menyumbang 76% data tetapi hanya ~9,5% yang berinvestasi; ada indikasi bahwa pesan produk belum menjawab “kecemasan finansial nyata” mereka.

  **Arah strategi:**

  - **Posisikan produk sebagai alat proteksi gaya hidup, bukan sekadar investasi**  
    - Contoh narasi:  
      - “Kalau terjadi apa-apa dengan pekerjaan, berapa bulan Bapak/Ibu bisa bertahan dengan tabungan sekarang?”  
      - “Produk ini membantu menyiapkan dana cadangan, bukan sekadar menambah risiko baru.”

  - **Segmentasi mikro dalam usia produktif**  
    - Bedakan pendekatan:
      - **29–35 tahun:** fokus pada mimpi (rumah pertama, pendidikan anak, mobil).  
      - **36–45 tahun:** fokus pada proteksi penghasilan & pendidikan anak.  
      - **46–52 tahun:** fokus pada kesiapan pensiun & mengurangi beban finansial di masa tua.

  - **Gunakan pendekatan bertahap (multi-touch)**  
    - Sentuhan 1: edukasi singkat (SMS/Email/WA) berisi fakta dan contoh.  
    - Sentuhan 2: telepon konsultatif, bukan “hard selling”.  
    - Sentuhan 3: follow-up dengan simulasi personal (misalnya disesuaikan dengan range gaji/tabungan).

---

#### 7.2.4 Penyesuaian pola interaksi untuk segmen under-performance

  1. **Batasi jumlah kontak berulang, tingkatkan kualitas kontak awal**  
    - Untuk segmen under-performance, tetapkan **batas kontak maksimal** (misal 5–7 kali).  
    - Pastikan **2–3 kontak pertama** sudah menggunakan skrip yang dioptimalkan (versi yang lebih sederhana dan relevan seperti di atas).

  2. **Variasikan konten setiap kali follow-up**  
    - Jangan mengulang skrip yang sama; setiap follow-up harus membawa “nilai baru”:  
      - info promo baru,  
      - contoh kasus baru,  
      - atau simulasi nominal yang berbeda.

  3. **Siapkan playbook skrip per persona**  
    - Buat template skrip terpisah untuk:
      - *Blue-collar middle school*  
      - *Services high school*  
      - *Usia 29–35 vs 36–52*  
    - Sertakan daftar **“objection & response”** yang paling sering muncul untuk tiap persona.

---

#### 7.2.5 . Inti rekomendasi

Alih-alih hanya mengurangi kontak ke segmen under-performance, Bank ABC perlu:

- **Mengubah cara berbicara** dengan mereka (bahasa, contoh, fokus manfaat).  
- **Mengubah urutan pesan** (bangun kepercayaan & relevansi dulu, baru bicara angka).  
- **Mengurangi tekanan “jualan produk”** dan memperbesar porsi peran bank sebagai **“mitra yang membantu mengamankan penghasilan dan masa depan”**.

Dengan strategi komunikasi yang lebih terarah ini, segmen under-performance diharapkan tidak lagi sekadar “beban volume kampanye”, namun dapat diubah menjadi **sumber konversi tambahan** secara bertahap dan terukur.


### 7.3 **Rekomendasi 3 – Evaluasi Tim Direct Marketing (Resource, Capability, dan Penanganan Volume Besar)**  

Berdasarkan analisis bulan (*`month`*) dan jumlah kontak kampanye (*`campaign`*), terlihat bahwa **volume kampanye yang sangat besar tidak sebanding dengan conversion rate yang didapatkan**. Pada bulan dan level campaign dengan jumlah kontak tinggi, tingkat konversi justru turun. Hal ini mengindikasikan bahwa **bottleneck utama ada pada kapasitas dan kualitas eksekusi tim direct marketing**, bukan semata pada produknya.

---

#### 7.3.1 Penataan ulang kapasitas & penjadwalan agent  

- **Hitung ulang beban kerja realistis per agent**  
  - Tetapkan batas **maksimum jumlah leads per agent per hari/bulan** yang masih memungkinkan penanganan berkualitas (durasi call memadai, ada waktu follow-up).  
  - Hindari “banjir leads” dalam satu periode tanpa penambahan resource atau redistribusi beban kerja.  

- **Sebar volume kampanye secara lebih merata**  
  - Ubah pola *single blast* menjadi **multi-wave campaign** (batch kecil bertahap) sehingga agent punya waktu cukup untuk membangun percakapan dan melakukan follow-up.  
  - Sinkronkan jadwal kampanye dengan **hari paling efektif** (Selasa–Kamis), bukan menumpuk di hari/pekan tertentu.  

---

#### 7.3.2 Penguatan capability penanganan leads  

- **Standarisasi playbook penanganan leads**  
  - Susun **alur baku**: opening → probing kebutuhan → edukasi produk → penawaran → penutupan + follow-up.  
  - Bedakan skrip untuk **segmen prioritas tinggi** dan **segmen under-performance** (selaras dengan Rekomendasi 1 & 2).  

- **Pelatihan ulang skill konsultatif agent**  
  - Fokus pada kemampuan menggali kebutuhan, mengelola keberatan, serta menjelaskan manfaat investasi secara sederhana dan relevan.  
  - Jadikan **agent dengan conversion tertinggi** sebagai *role model* / coach untuk sesi berbagi praktik terbaik.  

- **Tetapkan SLA follow-up yang jelas**  
  - Contoh: setiap nasabah yang menunjukkan minat harus difollow-up maksimal **X hari** sejak kontak pertama, bukan dibiarkan hilang karena agent keburu pindah ke batch leads berikutnya.  

---

#### 7.3.3 Strategi khusus untuk kampanye volume besar  

- **Pre-filter sebelum masuk ke agent**  
  - Lakukan **pre-screening** (segmentasi prioritas, saldo/gaji, riwayat respon) sehingga daftar leads yang masuk ke agent sudah relatif “hangat”.  
  - Leads dengan prioritas rendah dialihkan terlebih dahulu ke **channel massal** (SMS, push notification, email) sebelum di-*escalate* ke telemarketing.  

- **Prioritization rule di level agent**  
  - Di dalam *worklist* agent, terapkan **urutan prioritas otomatis** (mis. segmen prioritas tinggi + sinyal minat > segmen lain), sehingga energi agent fokus terlebih dahulu ke peluang dengan *return* tertinggi.  

- **Stop rule untuk kampanye yang tidak produktif**  
  - Jika conversion per 100 call pada segmen/skrip tertentu terus turun di bawah ambang batas, kampanye tersebut **dihentikan atau dirombak**, bukan dilanjutkan hanya demi menghabiskan daftar leads.  

---

#### 7.3.4 Monitoring kinerja & *continuous improvement*  

- **Bangun dashboard kinerja agent** dengan metrik:  
  - conversion per 100 call,  
  - rata-rata durasi call,  
  - jumlah follow-up yang tuntas,  
  - performa per channel (cellular vs telephone).  
  Metrik ini digunakan untuk **coaching berkala** dan penyesuaian target individu/tim.  

- **Quality assurance berbasis rekaman call**  
  - Lakukan sampling rekaman call untuk menilai **kualitas eksekusi skrip**, bukan hanya kuantitas panggilan.  
  - Integrasikan temuan QA ke dalam program **feedback & pelatihan berkala**.  

---

Dengan menjalankan rekomendasi ini, BANK ABC dapat mengurangi gap antara **volume kampanye** dan **konversi yang dihasilkan**, sehingga setiap kontak yang dilakukan memiliki probabilitas konversi lebih tinggi dan jam kerja tim direct marketing menjadi jauh lebih produktif.


### 7.4 **Rekomendasi 4 – Evaluasi Pendekatan Kontak dalam Eksekusi Kampanye (`campaign`)**  

Analisis variabel **campaign** menunjukkan bahwa **peningkatan jumlah kontak ke nasabah tidak selalu berbanding lurus dengan kenaikan konversi**.  

- **Kontak ke-1** memegang porsi nasabah terbesar (~17,6K) dan conversion rate tertinggi (**≈13%**).  
- **Kontak ke-2–3** masih cukup efektif (**≈11–10%**) dengan volume besar.  
- Mulai **kontak ke-4–7**, conversion turun ke **≈6–9%** dengan volume semakin sedikit.  
- Pada **kontak ≥ 8**, conversion rate banyak yang **mendekati 0%**, meski beberapa nasabah dihubungi lebih dari 10–20 kali.  

Temuan ini menegaskan adanya **diminishing return**: setelah titik tertentu, kontak tambahan hampir tidak menambah konversi dan justru menguras resource agent.

---

#### 7.4.1 Tetapkan *contact frequency cap* (batas jumlah kontak)

1. **Maksimalkan 1–3 kontak pertama**  
   - Perlakukan kontak ke-1 s.d. ke-3 sebagai **“prime window”** untuk konversi.  
   - Pastikan pada tiga kontak ini telah digunakan **skrip terbaik**, penjelasan produk yang jelas, dan *offer* yang paling menarik.  

2. **Batasi kontak telemarketing hingga 3–4 kali per kampanye**  
   - Setelah **3–4 kali gagal** (jawaban “tidak tertarik”, “sibuk”, atau tidak diangkat), nasabah dipindahkan ke:  
     - **channel pasif / massal** (SMS, email, push notification), atau  
     - *cooling period* sebelum diikutkan kampanye lain.  

3. **Stop rule untuk kontak berulang**  
   - Untuk nasabah dengan riwayat **≥ 8 kontak tanpa konversi**, hentikan telepon outbound untuk produk yang sama.  
   - Tag sebagai **“low-response / negative”** dan hanya diikutkan kembali jika ada *trigger* baru (produk berbeda, inbound interest, perubahan profil risiko).

---

#### 7.4.2 Diferensiasi pendekatan berdasarkan respons sebelumnya

1. **Berikan label/tag respons nasabah pada setiap kontak**  
   - Contoh: `prospect`, `not now`, `refuse`, `no answer`, `wrong number`.  
   - Gunakan tag ini untuk mengatur **jenis kontak berikutnya**:  
     - `prospect` → jadwalkan **follow-up terarah** (kirim materi dulu, lalu call).  
     - `not now` → jadwalkan **reminder** di bulan/periode yang lebih relevan.  
     - `refuse` → **jangan dihubungi untuk produk yang sama**.  

2. **Sesuaikan interval antar kontak**  
   - Hindari pola “dibombardir” berturut-turut (misal beberapa kali dalam minggu yang sama).  
   - Terapkan jeda yang wajar (mis. 2–4 minggu) agar nasabah tidak merasa terganggu dan agent punya waktu memproses leads lain.

---

#### 7.4.3 Prioritisasi kontak berdasarkan kualitas leads, bukan hanya kuantitas

1. **Urutkan daftar call berdasarkan skor prioritas**  
   - Gabungkan dimensi: **segmen prioritas (Rekomendasi 1)** + **riwayat respons positif** + **sinyal finansial**.  
   - Pastikan kontak ke-1–3 difokuskan pada **leads dengan potensi tertinggi**, bukan menyebar merata ke semua nasabah tanpa filter.

2. **Kurangi kontak berulang pada segmen under-performance**  
   - Untuk segmen seperti **middle.school, blue-collar, services** yang sudah menunjukkan conversion rendah,  
     - batasi jumlah *attempt*,  
     - dan utamakan channel non-voice (SMS/push) untuk sekadar *presence* produk.

---

#### 7.4.4 Monitoring efektivitas per level campaign

- Bangun ringkasan rutin yang menampilkan:  
  - **konversi per level campaign (1, 2, 3, …)**,  
  - **jumlah call** dan **konversi marginal** yang dihasilkan tiap kontak tambahan.  
- Gunakan laporan ini untuk:  
  - meninjau ulang **batas maksimum campaign**,  
  - menonaktifkan strategi yang ternyata hanya menambah call, bukan konversi.

---

Dengan pendekatan ini, BANK ABC dapat **mengurangi kontak yang tidak produktif**, menjaga pengalaman nasabah, serta mengalihkan waktu agent ke **kontak pertama–ketiga yang paling potensial menghasilkan konversi**, sehingga *ROI* kampanye direct marketing meningkat secara signifikan.


### 7.5 **Rekomendasi 5 – Optimasi Durasi Panggilan dan Kualitas Percakapan**

Berdasarkan analisis *duration*, terlihat bahwa **peluang konversi meningkat tajam pada panggilan dengan durasi ≥300 detik**, sementara panggilan sangat singkat (≤100 detik) hampir tidak menghasilkan penjualan. Temuan ini perlu diterjemahkan menjadi kebijakan operasional yang jelas.

---

#### 7.5.1. Prinsip umum durasi panggilan

- Jadikan **panggilan singkat (≤100 detik) sebagai “red flag”**:  
  banyak terjadi, tetapi conversion rate hanya ±0,8% → indikasi kuat call “asal kontak”.
- Dorong agent untuk membawa percakapan ke **rentang durasi efektif ≥300 detik**, di mana conversion rate mulai melonjak (±18,6% hingga ±48,6%).
- Fokus pada **kualitas dialog**, bukan sekadar memperpanjang waktu bicara.

---

#### 7.5.2. Kebijakan durasi dan desain skrip

1. **Tetapkan “durasi minimum efektif”**  
   - Misalnya, **targetkan sebagian besar call berada di rentang 300–600 detik** untuk leads yang menunjukkan minat.  
   - Panggilan yang selalu berakhir <100 detik perlu ditinjau (apakah karena skrip terlalu “jualan” atau cara opening kurang tepat).

2. **Perbaiki skrip agar mendorong dialog, bukan monolog**  
   - Tambahkan bagian *needs assessment* (tanya tujuan finansial nasabah, horizon waktu, toleransi risiko) di awal percakapan.  
   - Sisipkan pertanyaan terbuka yang memancing nasabah bercerita, sehingga durasi bertambah karena interaksi dua arah, bukan karena agent bicara sendiri.

3. **Gunakan skrip diferensiasi berdasarkan minat nasabah**  
   - Jika dalam 60–90 detik pertama nasabah menunjukkan minat (banyak bertanya, meminta penjelasan), agent diarahkan untuk **masuk ke mode konsultatif** dan siap mengalokasikan waktu lebih panjang (300–600+ detik).  
   - Jika nasabah jelas menolak atau tidak tertarik, agent boleh **mengakhiri call secara sopan** dan mencatat status, tanpa memaksakan durasi.

---

#### 7.5.3. Pengelolaan waktu agent dan pelatihan

- **Kalibrasi target produktivitas**:  
  - Jangan hanya menilai kinerja agent dari **jumlah call per hari**, tetapi kombinasikan dengan **proporsi call di durasi efektif** dan **conversion rate**.  
- **Pelatihan khusus “deep conversation”**:  
  - Latih agent dalam teknik membangun kepercayaan, mendengarkan aktif, dan menjawab keberatan, sehingga nyaman berada di percakapan 5–10 menit tanpa terasa memaksa.  

---

#### 7.5.4. Monitoring dan evaluasi berkala

- Pantau **distribusi `duration_cat` per agent** (≤100s, 101–300s, 301–600s, >600s) dan hubungkan dengan conversion rate masing-masing.
- Identifikasi **agent dengan kombinasi durasi seimbang dan konversi tinggi** sebagai *role model*, lalu gunakan *call recording* mereka sebagai bahan *coaching*.
- Secara berkala, lakukan **A/B testing skrip** untuk melihat apakah perubahan wording di awal percakapan dapat menggeser lebih banyak call ke durasi ≥300 detik.

---

Dengan mengelola **durasi panggilan sebagai salah satu tuas utama kualitas percakapan**, bank dapat mengurangi volume call yang tidak produktif (≤100 detik) dan memaksimalkan porsi interaksi yang benar-benar berpotensi closing, sehingga **rasio konversi meningkat tanpa harus menambah jumlah kontak secara berlebihan**.
